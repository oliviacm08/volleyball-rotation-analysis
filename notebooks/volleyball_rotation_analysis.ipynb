```python
import os
project_drive_path = '/content/drive/MyDrive/Volleyball_Analytics'
project_local_path = '/content/Volleyball_Analytics'

# Make sure local copy exists
if not os.path.exists(project_local_path):
    os.makedirs(project_local_path)
```

```python
import shutil
import os

# List all main folders in Drive
main_folders = ['scripts', 'data', 'notes', 'project_analysis']

# Define a function to ignore .gsheet files
def ignore_gsheet_files(directory, contents):
    return [c for c in contents if c.endswith('.gsheet')]

for folder in main_folders:
    src = os.path.join(project_drive_path, folder)
    dst = os.path.join(project_local_path, folder)
    if os.path.exists(src):
        # Use the ignore function to skip .gsheet files
        shutil.copytree(src, dst, ignore=ignore_gsheet_files, dirs_exist_ok=True)
```
